```python
import tkinter as tk
from tkinter import messagebox

# Create the main window
main = tk.Tk()
main.title("To-Do List App")
main.geometry("300x400")

# --- Widgets ---
# Label for entry field
task_label = tk.Label(main, text="Enter a task:", font=("Arial", 12, "bold"))
task_label.pack(pady=5)

# Entry widget for typing a task
task_entry = tk.Entry(main, width=30)
task_entry.pack(pady=5)

# Listbox to display tasks
task_listbox = tk.Listbox(main, width=40, height=10)
task_listbox.pack(pady=10)

# --- Functions ---
def add_task():
    """Adds the task from the entry box to the listbox."""
    task = task_entry.get()
    if task:
        task_listbox.insert(tk.END, task)
        task_entry.delete(0, tk.END)
    else:
        messagebox.showwarning("Input Error", "Please enter a task.")

def remove_task():
    """Removes the selected task from the listbox."""
    try:
        selected_index = task_listbox.curselection()
        task_listbox.delete(selected_index)
    except:
        messagebox.showwarning("Selection Error", "Please select a task to remove.")

# --- Buttons ---
# Button to add task
add_button = tk.Button(main, text="Add Task", command=add_task, bg="#4CAF50", fg="white")
add_button.pack(pady=5)

# Button to delete task
delete_button = tk.Button(main, text="Delete Task", command=remove_task, bg="#f44336", fg="white")
delete_button.pack(pady=5)

# Run the app
main.mainloop()
```
